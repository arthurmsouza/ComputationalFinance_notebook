```python
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from scipy.stats import mvn
import math
from matplotlib import collections  as mc
```

# Covariance matrix

```python
# Standard deviations and correlation
sig_x = 0.10
sig_y = 0.05
rho_xy = 0.9
# Covariance between X and Y
sig_xy = rho_xy * sig_x * sig_y
# Covariance matrix
Sigma_xy = x = np.array([sig_x**2, sig_xy, sig_xy, sig_y**2])
Sigma_xy.shape = (2,2)
print(Sigma_xy)
```

# Simulate data

```python
# Means
mu_x = 0.05
mu_y = 0.025

# Simulate 100 observations
random.seed(123)  # for reproducibility
xy_vals = np.random.multivariate_normal((mu_x, mu_y), Sigma_xy, 100)
print(xy_vals[0:10])
```

# Plot the simulated data

```python
%matplotlib inline
x,y=zip(*xy_vals)

fig = plt.figure()
plt.scatter(x,y)
fig.suptitle('Bivariate normal: rho = 0.9') 
plt.show()
```

# Add lines to the plot

```python
%matplotlib inline
x,y=zip(*xy_vals)

fig = plt.figure()
plt.scatter(x,y)
plt.axvline(x=mu_x)
plt.axhline(y=mu_y)
fig.suptitle('Bivariate normal: rho = 0.9') 
plt.show()
```

# Compute a joint probability

```python
%matplotlib inline
x,y=zip(*xy_vals)

fig, ax = plt.subplots(1, 1)
plt.scatter(x,y)
plt.axvline(x=mu_x)
plt.axhline(y=mu_y)

segments=[[(0, -1e10), (0, 0)],
      [(-1e10, 0), (0, 0)]]
lc = mc.LineCollection(segments, linewidths=2, color='r')
ax.add_collection(lc)
      
fig.suptitle('Bivariate normal: rho = 0.9') 
plt.show()

# Compute joint probability
low = np.array([-999999, -999999])
upp = np.array([0, 0])
mu = np.array([mu_x, mu_y])
p,i = mvn.mvnun(low,upp,mu,Sigma_xy)
print (p)
```
